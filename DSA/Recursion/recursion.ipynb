# Recursion:
function calling itself is called recursion

```python
def fun1(n):
    if n<=0:
        return
    print(f"Fun1 is called {n} times")
    fun1(n-1)

fun1(5)

# def fun2():
#     print("before fun1")
#     fun1()
#     print("after fun1")
```

### Application of Recursion
    * Dynamic Programming
    * Backtracking 
    * Divide & conquer(binary search, quick sort & merge sort)
    


```python
#factorial
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)

print(fact(5))
```

```python
#fibonacci 

def fib(n):
    if n==0:
        return 0
    if n==1:
        return 1
    
    return fib(n-1) + fib(n-2)

print(fib(10))
```

```python
#tail recursion

def fun3(n):
    if n<=0:
        return
    print(n, end="")
    fun3(n-1)

fun3(6)
```

```python
#recursion practice 1

def fun(n):
    if n==0:
        return
    print(n)
    fun(n-1)
    print(n)
fun(3)


```

```python
#recursion practice 2

def fun(n):
    if n==0:
        return
    fun(n-1)
    print(n)
    fun(n-1)
fun(3)
```

```python
def fun(n):
    if n <= 1:
        return 0
    else:
        return 1+fun(n/2)

fun(16)
print(fun(20))
```

```python
def fun3(n):
    if n==0:
        return
    fun3(n/2)
    print(n%2)

fun3(13)
```

```python
#recursive method

def print1toN(n):
    if n==0:
        return
    print1toN(n-1)
    print(n)




print1toN(4)
```

```python
#sum of digits recursion

def sumOfdig(n):
    if n<10:
        return n
    return sumOfdig(n//10) + n%10

sumOfdig(984)
```

```python
#palindrone
def checkPalindrome(st, start, end):
    print(st, start, end)
    if start>=end:
        return True
    return(st[start]==st[end] and checkPalindrome(st, start+1,end-1))
    
Inp = "abbanabba"
print(checkPalindrome(Inp, 0, len(Inp)-1))    

```

```python
def checkPal(st):
    print(st)
    l = len(st)

    if l<2:
        return True
    
    elif st[0]==st[l-1]:
        return checkPal(st[1:l-1])
    
    else:
        return False
Inp = "abbdabba"
print(checkPal(Inp) ) 
    
```
